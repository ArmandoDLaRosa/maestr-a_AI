Recuerda que:


*   Una **clave primaria** es un atributo (o conjunto) que identifica unívocamente a cada registro en la tabla.
*   Una **clave foránea** (externa o ajena) es un atributos (o conjunto) en una tabla que es una clave primaria en otra (o posiblemente la misma) tabla.
*   Las **relaciones** son las líneas que conectan una tabla con otra y el extremo determina la cardinalidad. Las relaciones con línea continua (identificadora) representan una transformación donde la clave primaria de una tabla pasa a ser foránea y primaria (al mismo tiempo) de otra. Las relaciones con línea discontinua (no identificadora) representan una transformación donde la clave primaria de una tabla pasa a ser sólo foránea en otra.

```python
import sqlalchemy as sqla
import pymysql
import pandas as pd
```

1.	Crea el motor `sqlalchemy`, con el método `create_engine()` y una conexión con `connect()` como se muestra a continuación:

```python
# Crear el motor (dialecto://usuarioBD:clave@ipHostDBMS:puerto/esquema
db = sqla.create_engine('mysql+pymysql://:!@/classicmodels', pool_recycle=3600)

# Crea una conexión para luego invocar declaraciones SQL
conn = db.connect()
```

Escribe las consultas en SQL para obtener:

2.	La información de las líneas de productos.

```python
pd.read_sql("SELECT * FROM productlines;", conn)
```

3.	Los empleados ordenados por nombre.

```python
pd.read_sql("""
            SELECT * 
            FROM employees 
            ORDER BY firstName, lastName;""", conn)
```

4.	Los países donde hay oficinas.

```python
pd.read_sql("SELECT DISTINCT country FROM offices;", conn)
```

5.	El nombre y teléfono de los clientes de Nueva York (*NYC*).

```python
pd.read_sql("""SELECT customerName, phone 
               FROM customers 
               WHERE city = 'NYC';""", conn)
```

6.	El código y nombre de los productos del vendedor *Gearbox Collectibles* que tengan menos de 1000 unidades en stock.

```python
pd.read_sql("""SELECT productCode, productName FROM products WHERE productVendor = 'Gearbox Collectibles' AND quantityInStock < 1000;""", conn)
```

7.	Los tres productos más caros, desde el punto de visto de los comercializadores (`buyPrice`).

```python
pd.read_sql("SELECT productCode, productName, buyPrice FROM products ORDER BY buyPrice DESC LIMIT 3;", conn)
```

8.	La cantidad de productos por línea de producto (no las existencias en inventario)

```python
pd.read_sql("SELECT productLine, COUNT(productCode) AS cantidadProducto FROM products GROUP BY productLine;", conn)
```

9.	La cantidad de empleados por país (tomando en cuenta la ubicación de la oficina).

```python
pd.read_sql("""SELECT o.country, COUNT(e.employeeNumber) AS ConteoEmpleado 
               FROM offices o 
               JOIN employees e ON o.officeCode = e.officeCode 
               GROUP BY o.country;""", conn)
```

10.	El promedio de los pagos de cada uno de los clientes de España.

```python
pd.read_sql("""
            SELECT c.customerNumber, c.customerName, AVG(p.amount) AS promedioPago 
            FROM customers c 
            JOIN payments p ON c.customerNumber = p.customerNumber 
            WHERE c.country = 'Spain' 
            GROUP BY c.customerNumber, c.customerName;""", conn)
```

# **Parte 2**. Manipulación de datos con Pandas

11.	Carga las tablas empleadas en dataframes con el mismo nombre y resuelve las consultas anteriores con las funciones de Pandas. Cuida no sobreescribir los dataframes originales al resolver las consultas. Debes obtener los mismos resultados que con SQL.

```python
customers = pd.read_sql("SELECT * FROM customers", conn)
employees = pd.read_sql("SELECT * FROM employees", conn)
offices = pd.read_sql("SELECT * FROM offices", conn)
productlines = pd.read_sql("SELECT * FROM productlines", conn)
products = pd.read_sql("SELECT * FROM products", conn)
orderdetails = pd.read_sql("SELECT * FROM orderdetails", conn)
orders = pd.read_sql("SELECT * FROM orders", conn)
payments = pd.read_sql("SELECT * FROM payments", conn)
```

```python
# Pregunta 1  La información de las líneas de productos.
productlines.copy()
```

```python
# Pregunta 2 Los empleados ordenados por nombre.
employees.sort_values(by=['firstName', 'lastName'])
```

```python
# Pregunta 3  Los países donde hay oficinas.
offices['country'].unique()
```

```python
# Pregunta 4  Los clientes de la ciudad de Nueva York.
customers[['customerName', 'phone']][customers['city'] == 'NYC']
```

```python
# Pregunta 5  Los productos de la marca Gearbox Collectibles y que tengan menos de 1000 unidades en stock.
products[['productCode', 'productName']][(products['productVendor'] == 'Gearbox Collectibles') & (products['quantityInStock'] < 1000)]
```

```python
# Pregunta 6  Los 3 productos más caros.
products[['productCode', 'productName', 'buyPrice']].sort_values(by='buyPrice', ascending=False).head(3)
```

```python
# Pregunta 7 La cantidad de productos por línea de producto.
products.groupby('productLine').size().reset_index(name='productCount')
```

```python
# Pregunta 8 La cantidad de empleados por país.
merged_df = pd.merge(employees, offices, on='officeCode')
merged_df.groupby('country').size().reset_index(name='employeeCount')
```

```python
# Pregunta 9 El promedio de pagos por cliente de España.
merged_df = pd.merge(customers[customers['country'] == 'Spain'], payments, on='customerNumber')
merged_df.groupby(['customerNumber', 'customerName'])['amount'].mean().reset_index(name='averagePayment')
```
